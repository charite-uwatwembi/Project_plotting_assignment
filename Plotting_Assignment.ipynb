```python

import seaborn as sns
import matplotlib.pyplot as plt
Co2_emission = df['co2']
sns.set_style("whitegrid")

sns.histplot(Co2_emission, bins=20, color='pink', edgecolor='red')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide Emissions')
plt.show()
print(Co2_emission.head(10))
```

**Interpretaion**
This histogram visually represents how CO2 emissions are distributed within the dataset, offering insights into the central tendency, spread, and potential outliers of the emissions data.

```python
# YOUR CODE HERE

Displacement = df['displ']
sns.set(style="whitegrid")
sns.histplot(data=df, x='displ', color='green', edgecolor='red', bins=20)
plt.xlabel('Engine Displacement')
plt.ylabel('Frequency')
plt.title('Distribution of Engine Displacement')
plt.show()
print(Displacement.head(10))
```

**Interpratation**
The histogram illustrates the frequency distribution of engine displacement values in the dataset. Engine displacement, measured in volume of air displaced by engine pistons, is shown on the x-axis, while the frequency of occurrence for each displacement value is depicted on the y-axis. This visualization provides insights into the prevalence and distribution pattern of engine displacement values, aiding in understanding the dataset's typical range and identifying any notable trends or outliers.

```python
# YOUR CODE HERE
Combfuel= df['comb']
sns.histplot(data=df, x='comb', color='black', edgecolor='red', bins=20)
plt.xlabel('Combined Fuel Efficiency')
plt.ylabel('Frequency')
plt.title('Distribution of Combined Fuel Efficiency')
plt.show()
print(Combfuel.head(10))
```

**Interpretation :**
the histogram illustrates the distribution of combined fuel efficiency values from a dataset, showing how often different efficiency levels occur. The choice of 20 bins suggests a moderate level of detail, allowing for a clear overview of the data's spread without overwhelming the viewer with too much information. The black and red color scheme, along with the clear labeling and title, make the histogram easy to understand and interpret.

```python
numbercolumns = df.select_dtypes(include=['number'])

corr_matrix = numbercolumns.corr()


plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Fuel Economy Data')

plt.show()
```

**Interpretation**
the heatmap illustrates the correlation matrix of numerical variables in a dataset related to fuel economy, including specific columns that are numerical in nature. It shows how each variable is related to every other variable, with the color and the correlation coefficient indicating the strength and direction of these relationships. This visualization is particularly useful for identifying which variables are strongly correlated, which can inform further analysis or decision-making processes.
